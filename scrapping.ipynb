Importo Librerías

```python
import pandas as pd
import yfinance as yf
```

Descargo los datos desde el año 2000 hasta la actualidad

```python
Nvidia = yf.download('NVDA', start='2000-01-01', end='2023-01-31', rounding=True, group_by='Date')
Nvidia
```

```python
AMD = yf.download('AMD', start='2000-01-01', end='2023-01-31')
AMD
```

Exporto los datos en un csv

```python
Nvidia.to_csv('Nvidia.csv')
```

```python
AMD.to_csv('AMD.csv')
```
